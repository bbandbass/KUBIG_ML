#Preprocessing

```python
X = mushroom_df.drop('class', axis = 1)
y = mushroom_df['class']
X.head()
```

```python
y.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in X.columns:
  X[col] = le.fit_transform(X[col])
y = le.fit_transform(y)
```

```python
X.head()
```

```python
y
```

1 for poisonous mushrooms and 0 for eatable mushrooms.

```python
X = pd.get_dummies(X, columns = X.columns, drop_first = True)
```

```python
X.head()
```

#Train/Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3,  stratify = y)
```

```python
X_train
```

#Model Selection

##Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, precision_score, recall_score
from sklearn.metrics import confusion_matrix, precision_recall_curve, roc_curve, auc, log_loss
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
y_pred_proba = logreg.predict_proba(X_test)[:, 1]

[fpr, tpr, thr] = roc_curve(y_test, y_pred_proba)

print('Train/Test split results:')
print('LogisticRegression' + ' accuracy is %2.3f' %accuracy_score(y_test, y_pred))
```

```python
cnf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(pd.DataFrame(cnf_matrix), annot = True, fmt = 'f')
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('Actual Label')
```

LogisticRegression has accuracy of about 100%. But not technically 100% because there are 1 FN value according to the confusion matrix.

##KNN Classifier

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

sc = StandardScaler()
X2 = mushroom_df.drop('class', axis = 1)
y2 = mushroom_df['class']

for col in X2.columns:
  X2[col] = le.fit_transform(X2[col])
y2 = le.fit_transform(y2)

X2 = sc.fit_transform(X2)
X2
```

```python
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size = 0.3, stratify = y)
```

```python
test_scores = []
train_scores = []

for i in range(1, 15):
  
  knn = KNeighborsClassifier(i)
  knn.fit(X2_train, y2_train)

  train_scores.append(knn.score(X2_train, y2_train))
  test_scores.append(knn.score(X2_test, y2_test))
```

```python
print(train_scores)
print(test_scores)
```

```python
max_train_score = max(train_scores)
train_scores_ind = [i for i, v in enumerate(train_scores) if v == max_train_score]
print('Max train scores {} % and k = {}'.format(max_train_score * 100, list(map(lambda x: x + 1, train_scores_ind))))
```

```python
max_test_score = max(test_scores)
test_scores_ind = [i for i, v in enumerate(test_scores) if v == max_test_score]
print('Max test score {} % and k = {}'.format(max_test_score * 100, list(map(lambda x: x + 1, test_scores_ind))))
```

```python
knn = KNeighborsClassifier(1)
knn.fit(X2_train, y2_train)
y2_pred = knn.predict(X2_test)
cnf_matrix = confusion_matrix(y2_test, y2_pred)
sns.heatmap(pd.DataFrame(cnf_matrix), annot = True, fmt = 'f')
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('Actual Label')
```

KNN Classification has accuracy of 100% as well with the best KNN values k = 1, 2, 3, 4, 5, 6, 7, 9, 10 and 11.
